```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler #balance the clustered columns
```

```python
df=pd.read_csv("/content/customers.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#drop customer id because it is a unique column
df.drop(["CustomerID"],axis=1,inplace=True)
```

```python
#feature extraction/feature selection
x=df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
#scale the data(balance the data)
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
#initialize the KMeans
km=KMeans(n_clusters=5,random_state=42)#control the randomness of the ai
```

```python
#apply the algorithm on the data
df['cluster']=km.fit_predict(x)#assign the clusters to the cluster column
```

```python
#visualize the clusters
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'],c=df['cluster'])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title("Kmeans Clustering of Annual Income vs Spending Score")
plt.show()
```
